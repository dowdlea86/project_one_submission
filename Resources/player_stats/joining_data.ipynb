```python
import pandas as pd
from pathlib import Path
```

```python
# file path
standard_stats_path = Path("standard_stats.csv")
defensive_stats_path = Path("defensive_actions.csv")
#goalkeeping_stats_path = Path("goalkeeping.csv")
```

```python
#read in data
standard_stats = pd.read_csv("standard_stats.csv", encoding = "ISO-8859-1", engine='python')
defensive_stats = pd.read_csv("defensive_actions.csv", encoding = "ISO-8859-1", engine='python')
defensive_stats = defensive_stats.rename(columns={
    "ï»¿player": "player"
})
defensive_stats.head()
```

```python
#join first 2 stats
def_and_standard = standard_stats.merge(defensive_stats, how='left', on=['player', 'year'])
def_and_standard.tail()
```

```python
#create path for passing
passing_stats_path = Path("passing.csv")
passing_stats = pd.read_csv("passing.csv", encoding = "ISO-8859-1", engine='python')
passing_stats.head()
```

```python
# combine data
def_stand_pass = def_and_standard.merge(passing_stats, how='left', on=['player', 'year'])
def_stand_pass.tail()
```

```python
#shooting stats
shooting_stats_path = Path("player_shooting.csv")
shooting_stats = pd.read_csv("player_shooting.csv", encoding = "ISO-8859-1", engine='python')
shooting_stats.head()
```

```python
#merge data
non_gk_stats = def_stand_pass.merge(shooting_stats, how='left', on=['player', 'year'])
non_gk_stats.tail()
```

```python
#cleaning data
non_gk_stats = non_gk_stats.dropna()
non_gk_stats.head()
```

```python
#updating the year column
non_gk_stats['year'] = non_gk_stats['year'].astype('int')
non_gk_stats.head()
```

```python
non_gk_stats['born_x'] = non_gk_stats['born_x'].astype('int')
non_gk_stats.head()
```

```python
#output to csv

non_gk_stats.to_csv(r'non_gk_stats.csv', index = False)
```
